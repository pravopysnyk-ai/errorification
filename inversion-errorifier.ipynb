```python
import random

def switch_words(sentence):
    words = sentence.split()
    n = len(words)
    random_word_index = random.randint(0, n-1)
    for i in range(max(0, random_word_index - 2), min(n-1, random_word_index + 3)):
        if i != random_word_index:
            words[i], words[random_word_index] = words[random_word_index], words[i]
            break
    return ' '.join(words)
```

```python
# clean data FILE to be errorified
input_file = "/content/drive/MyDrive/artem-yushko/data-artem/cleaned/borshch4.txt"

# output FOLDER for the errorified and tagged data, future model input
out_folder = "/content/drive/MyDrive/UNLP/assist-data/inversion-errors-10k"
```

```python
import os
# creating the output folder
if not os.path.exists(out_folder):
  os.mkdir(out_folder)

# reading the file
with open(input_file, 'r') as f:
  text = f.read()
  lines = text.split('\n')
```

```python
dataset = lines[:10000]
```

```python
out_sentences = []
for sentence in dataset:
  errorified = switch_words(sentence)
  out_sentences.append(errorified)
```

```python
text = '\n'.join(out_sentences)
with open(out_folder + "/source.txt", 'w') as f:
  f.writelines(text)

text = '\n'.join(dataset)
with open(out_folder + "/target.txt", 'w') as f:
  f.write(text)
```
